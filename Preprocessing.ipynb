# Importing the libraries and Raw Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('Raw_Data.xlsx')
```

# Data Preprocessing and Creating features

```python
import re
from re import search
Team = []
Player_Name = []
IsCaptain = []
Playing_Role = []
Age = []
Batting_style= []
Bowling_style = []
IsOverseas = []
IsUncapped= []

pattern = re.compile("Age:")

all_teams = df.columns.tolist()

for team in all_teams:
    for i,j in df[team].dropna().items():
        if search(pattern, j):
            j = j.replace("Playing",",Playing").replace("Batting:", ",Batsman:").replace("Bowling:", ",Bowling:").replace('\xa0',"")
            Team.append(team)
            Player_Name.append(df[team][i-1].replace('Captain/Keeper',"").replace("Captain","").replace('Overseas Player',"").replace('Wicketkeeper', ""))
            Age.append(int(j.split(':')[1].split()[0]) + int(j.split(':')[1].split()[2])/365)
            Playing_Role.append(j.split(':')[2].split(',')[0])
            Batting_style.append(j.split(':')[3].split(',')[0])
            try:
                Bowling_style.append(j.split(':')[4].split(',')[0])
            except IndexError:
                Bowling_style.append('nan')

            if search('Overseas Player', df[team][i-1]):
                IsOverseas.append('Yes')
            else:
                IsOverseas.append('No')

            if search('Captain', df[team][i-1]):
                IsCaptain.append('Yes')
            else:
                IsCaptain.append('No')
        
        
        #Age.append(re.search('[\d,\S,\w,\S,\d]+ days',j))
(Player_Name)
```

# Creating the data frame from above results

```python
data = {'Team': Team, 'Player_Name': Player_Name, 'IsCaptain': IsCaptain, 'Age': Age, 'Playing_Role': Playing_Role, 'Batting_style': Batting_style, 'Bowling_style': Bowling_style, 'IsOverseasPlayer': IsOverseas}
Main_df = pd.DataFrame(data, columns = data.keys())
Main_df.head()
```
